```python
#Média aritmética
nota1 = input("'Digite a nota do trabalho ").split()
nota2 = input("'Digite a nota do 2 trabalho ").split()

nota1 = [float(valor) for valor in nota1]
nota2 = [float(valor) for valor in nota2]

# Calculando a média da primeira variável
soma_nota1 = 0
contador_nota1 = 0

for valor in nota1:
    soma_nota1 += valor
    contador_nota1 += 1

if contador_nota1 != 0:
    media_nota1 = soma_nota1/ contador_nota1
else:
    media_nota1 = 0

# Calculando a média da segunda variável
soma_nota2 = 0
contador_nota2 = 0

for valor in nota2:
    soma_nota2 += valor
    contador_nota2 += 1

if contador_nota2 != 0:
    media_nota2 = soma_nota2 / contador_nota2
else:
    media_nota2 = 0

print("Média da primeira nota:", media_nota1)
print("Média da segunda nota:", media_nota2)
```

```python
#Moda (se houver)
nota1 = float(input("Digite o primeiro valor: "))
nota2 = float(input("Digite o segundo valor: "))

# Criar uma lista com os valores fornecidos
notas = [nota1, nota2]

# Criar um dicionário para contar a frequência dos valores
frequencia = {}
for nota in notas:
    if nota in frequencia:
        frequencia[nota] += 1
    else:
        frequencia[nota] = 1

# Encontrar o valor com maior frequência (moda)
moda = None
max_frequencia = 0
for nota, freq in frequencia.items():
    if freq > max_frequencia:
        moda = nota
        max_frequencia = freq

if max_frequencia > 1:
    print("A moda das notas é:", moda)
else:
    print("Não há moda para as notas fornecidos.")
```

```python
#Variância amostral
nota1 = float(input("Digite o primeiro valor: "))
nota2 = float(input("Digite o segundo valor: "))

notas = [nota1, nota2]

media = sum(notas) / len(notas)   # Calcular a média dos valores

variancia = sum((x - media) ** 2 for x in notas) / len(notas)   # Calcular a variância amostral
print("A variância amostral das notas é:", variancia)
```

```python
#Q1
nota1 = float(input("Digite o primeiro valor: "))
nota2 = float(input("Digite o segundo valor: "))
notas = [nota1, nota2]

indice_q1 = int(len(notas) / 4)

if len(notas) % 4 == 0:
    q1 = (notas[indice_q1 - 1] + notas[indice_q1]) / 2
else:
    q1 = notas[indice_q1]
print("O primeiro quartil é:", q1)
```

```python
#Calculando Q3 (terceiro quartil)
nota1 = float(input("Digite a nota do trabalho 1: "))
nota2 = float(input("Digite a nota do trabalho 2: "))
notas = [nota1, nota2]

indice_q3 = int(3 * len(notas) / 4)

if len(notas) % 4 == 0:
    q3 = (notas[indice_q3 - 1] + notas[indice_q3]) / 2 # Se o índice for par, calcular a média dos valores do índice e do índice - 1
else:
    q3 = notas[indice_q3] # Se o índice for ímpar, o valor do índice representa o terceiro quartil
print("O terceiro quartil é:", q3)
```

```python
#AIQ (amplitude interquartílica)

nota1 = float(input("Digite a nota do trabalho 1: "))
nota2 = float(input("Digite a nota do trabalho 2: "))
notas=[nota1,nota2]

indice_q1 = int(len(notas) / 4)

# Calculando o índice do terceiro quartil (Q3)
indice_q3 = int(3 * len(notas) / 4)

# Calculando o primeiro quarrtil (Q1)
if len(notas) % 4 == 0:
    q1 = (notas[indice_q1 - 1] + notas[indice_q1]) / 2
else:
    q1 = notas[indice_q1]

# Calculando o terceiro quartil (Q3)
if len(notas) % 4 == 0:
    q3 = (notas[indice_q3 - 1] + notas[indice_q3]) / 2
else:
    q3 = notas[indice_q3]

amplitude_interquartilica = q3 - q1

# Exibir o resultado
print("A amplitude interquartílica de notas é:", amplitude_interquartilica)
```
